## 4.Follow/Unfollow given handle - 
1.Open the Instagram Handle of “So Delhi”
2.Start following it. Print a message if you are already following
3.After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
#function for following a certain account
def following(username):
    search=driver.find_element_by_xpath('//input[contains(@class,"XTCLo")]')  #searching the handle in the search box
    search.send_keys(username)
    
    link=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]/a')))  #wait until the handle is visible

    link.click()  #opening the handle
    follow=wait.until(EC.presence_of_element_located((By.XPATH,'//button[contains(@class,"_5f5mN")]')))  #waiting for the follow button to appear
    
    if follow.text=='Follow':  #checking if the account has been followed already
        print("not following")  #if not then click the follow button
        follow.click()
    else:
        print("following")  #message if the accont is already following

#funtion for unfollowing an account
def unfolowing(username):
    search=driver.find_element_by_xpath('//input[contains(@class,"XTCLo")]')
    search.send_keys(username)
    
    link=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]/a')))

    link.click()
    follow=wait.until(EC.presence_of_element_located((By.XPATH,'//button[contains(@class,"_5f5mN")]')))
    
    if follow.text!='Follow':
        print("following")
        follow.click()
        unfollow=driver.find_element_by_xpath('//button[contains(@class,"aOOlW")]')
        unfollow.click()
    else:
        print("not following")
```

```python
following('So Delhi')  #the calling statement
```

## 5. Like/Unlike posts
1.Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
2.Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
#function for liking a certain number of posts 
def like_post(str):
    driver.get('https://www.instagram.com/' + str)  #opening the profile
    time.sleep(2)
    img_url=wait.until(EC.driver.find_element_by_class_name('_9AhH0')   #opening the first post
    img_url.click()
    time.sleep(2)
    for i in range(30):  #liking the post givn in the range
        like=driver.find_element_by_xpath('//span/button[@class="wpO6b "]') #searching for the like button
        like.click()  #clicking the like button
        time.sleep(1)
        next_pic=driver.find_element_by_xpath('//div[@class="DdSX2"]/a[contains(@class,"coreSpriteRightPaginationArrow")]')
        next_pic.click()
        time.sleep(2)                   
```

## 6. Extract list of followers
1. Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
2. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
#function for extracting the list of followers 
def followers(str):
    driver.get('https://www.instagram.com/' + str)  #opening the pro
    time.sleep(3)
    li=driver.find_elements_by_class_name('-nal3')    ##the follower button
    li[1].click()
    time.sleep(2)
    curr_ht=driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')  #current position  of the scroll
    #new_ht=0
    tot_followers=set()
    while True:
        followerlist=driver.find_elements_by_xpath('//a[contains(@class,"FPmhX")]') #list of followers obtained
        for i in followerlist:
            print(i.text)
            
            if len(tot_followers)<70:
                #print(len(tot_followers))
                tot_followers.add(i.text)
            else:
                break
        if len(tot_followers)<70 :
            #print(len(tot_followers))
            driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0,arguments[0])',curr_ht)  #scrolling
        else:
            break
        new_ht=driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;') #new height of the scroll
        curr_ht=new_ht 
        #print(len(tot_followers))

    print(len(tot_followers))  #length of total followers
    return tot_followers   #returing the list of followers
```

```python
tot_followers=followers('foodtalkindia')  #calling the function for foodtalkindia
```
